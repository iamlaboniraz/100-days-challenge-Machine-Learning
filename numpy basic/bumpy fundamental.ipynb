## we can create array using numpy array

```python
import numpy as np

a = np.array([1,2,3])
print()
```

```python
b = np.array([[1,2,3],[4,5,6]])
print(b)
```

```python
np.array([1,2,3],dtype=float)
```

```python
np.array([1,2,3],dtype=complex)
```

```python
# np.arange 
np.arange(1,11,2) # as like python range function
```

```python
# with reshape
np.arange(1,11).reshape(2,5)
```

```python
# np.ones and np.zeros
np.ones((3,4))
```

```python
np.zeros((3,4))
```

```python
np.random.random((3,4))
```

```python
# np.linspace
np.linspace(-10,10,10)
```

```python
# np.identity
np.identity(3)
```
