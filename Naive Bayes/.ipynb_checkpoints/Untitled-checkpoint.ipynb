```python
# Solution
# P(Yes | Sunny,hot,high,weak) = P(Sunny| Yes) * P(Hot|Yes) * P(High| Yes) * P(Weak | Yes) * P(Yes)
# P(No | Sunny,hot,high,weak) = P(Sunny| No) * P(Hot|No) * P(High| No) * P(Weak | No) * P(No)
# compare and decide using the Maximum a posteriori Rule
```
