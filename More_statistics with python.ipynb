```python
data = pd.read_csv("data.csv")
```

```python
data = data.drop(['Unnamed: 32','id'], axis=1)
```

```python
m = plt.hist(data[data['diagnosis']=='M'].radius_mean, bins=30, fc=(1,0.5,0,0.5), label='Malignant')
b = plt.hist(data[data['diagnosis']=='B'].radius_mean, bins=30, fc=(1,0,0,0.5), label='Bening')
plt.legend()
plt.xlabel("Redius Mean Values")
plt.ylabel("Frequency")
plt.show()
```

```python
frequent_malignant_radius_mean = m[0].max()
frequent_malignant_radius_mean
```

```python
index_frequent_malignant_radius_mean = list(m[0]).index(frequent_malignant_radius_mean)
index_frequent_malignant_radius_mean
```

```python
most_frequent_malignant_radius_mean = m[1][index_frequent_malignant_radius_mean]
most_frequent_malignant_radius_mean
```

```python
most_frequent_malignant_radius_mean = m[1][index_frequent_malignant_radius_mean]
print("Most frequent malignant radius mean is: ",most_frequent_malignant_radius_mean)
```
