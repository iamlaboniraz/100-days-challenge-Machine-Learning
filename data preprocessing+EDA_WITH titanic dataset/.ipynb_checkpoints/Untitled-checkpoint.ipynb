## Wrangle data
- Based on our assumptions and decision we can drop `Cabin` and `Ticket` fetures

```python
print('Before', train_df.shape, test_df.shape, combine[0].shape, combine[1].shape)

train_df = train_df.drop(['Cabin','Ticket'],axis=1)
test_df = test_df.drop(['Cabin','Ticket'],axis=1)
combine = [train_df,test_df]

print('After', train_df.shape, test_df.shape, combine[0].shape, combine[1].shape)
```

## Creating new feature extracting from existing

```python
for dataset in combine:
    dataset['Title'] = dataset.Name.str.extract(' ([A-Za-z]+)\.',expand=False)
pd.crosstab(train_df['Title'],train_df['Sex'])
```

```python
for dataset in combine:
    dataset['Title'] = dataset['Title'].replace(['Capt','Col','Countess','Don','Dr','Jonkheer',
                                               'Lady','Major','Rev','Sir'],'Rare')
    dataset['Title'] = dataset['Title'].replace('Mlle','Miss')
    dataset['Title'] = dataset['Title'].replace('Ms','Miss')
    dataset['Title'] = dataset['Title'].replace('Mme','Mrs')
train_df[['Title','Survived']].groupby(['Title'],as_index=False).mean()
```

```python
title_mapping = {'Mr':1,'Miss':2,'Mrs':3,'Master':4,'Rare':5}
for dataset in combine:
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title']= dataset['Title'].fillna(0)
train_df['Title'].value_counts()
```

```python
train_df.head()
```

- Now we can drop `Name` feature into our train_df and test_df dataset and also `PassengerId` not needed in training dataset

```python
train_df = train_df.drop(['Name','PassengerId'],axis=1)
test_df = test_df.drop(['Name'],axis=1)
combine = [train_df, test_df]
```

```python
train_df.shape, test_df.shape
```

## Converting a categorical feature

```python
for dataset in combine:
    dataset['Sex'] = dataset['Sex'].map({'female':1,'male':'0'}).astype(int)
train_df.head()
```
